```python
from scipy import stats
df = df[columnimportantes][(np.abs(stats.zscore(df[columnimportantes]['SalePrice'])) < 3)]
sns.boxplot(x=df[columnimportantes]['SalePrice'])
```

```python
df = df[(np.abs(stats.zscore(df['TotalBsmtSF'])) < 3)]
sns.boxplot(x=df[columnimportantes]['TotalBsmtSF'])
```

```python
# Eliminar valores atípicos en la variable GrLivArea
Q1 = df['GrLivArea'].quantile(0.25)
Q3 = df['GrLivArea'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['GrLivArea'] >= Q1 - 1.5 * IQR) & (df['GrLivArea'] <= Q3 + 1.5 * IQR)]
sns.boxplot(x=df['GrLivArea'])
```

```python
df = df[(np.abs(stats.zscore(df['YearBuilt'])) < 3)]
sns.boxplot(x=df[columnimportantes]['YearBuilt'])
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
#min-max
# Aplicamos el escalado Z-score (usando StandardScaler)
df[['SalePrice_scaled', 'GrLivArea_scaled', 'YearBuilt_scaled', 'TotalBsmtSF_scaled', 
    'GarageCars_scaled', 'FullBath_scaled', 'OverallQual_scaled']] = scaler.fit_transform(
    df[['SalePrice', 'GrLivArea', 'YearBuilt', 'TotalBsmtSF', 'GarageCars', 'FullBath', 'OverallQual']]
)

# Aplicamos el escalado Min-Max (usando MinMaxScaler)
df[['SalePrice_normalized', 'GrLivArea_normalized', 'YearBuilt_normalized', 'TotalBsmtSF_normalized',
    'GarageCars_normalized', 'FullBath_normalized', 'OverallQual_normalized']] = min_max_scaler.fit_transform(
    df[['SalePrice', 'GrLivArea', 'YearBuilt', 'TotalBsmtSF', 'GarageCars', 'FullBath', 'OverallQual']]
)

print(df[['SalePrice', 'SalePrice_scaled', 'SalePrice_normalized', 
          'GrLivArea', 'GrLivArea_scaled', 'GrLivArea_normalized', 
          'YearBuilt', 'YearBuilt_scaled', 'YearBuilt_normalized', 
          'TotalBsmtSF', 'TotalBsmtSF_scaled', 'TotalBsmtSF_normalized',
          'GarageCars', 'GarageCars_scaled', 'GarageCars_normalized',
          'FullBath', 'FullBath_scaled', 'FullBath_normalized',
          'OverallQual', 'OverallQual_scaled', 'OverallQual_normalized']].head())
```

```python
#Verficando el sesgo de las columnas numericas
skewness = df[columnimportantes].apply(lambda x: stats.skew(x.dropna()))
print(f'\nSesgo de las columnas importantes:\n{skewness}')
```

```python
#Verificando el sesgo mas alto entre de las columnas numericas
skewed_columns = skewness[skewness.abs() > 0.5].index
print(f'\nColumnas con sesgo alto:\n{skewed_columns}')
```

```python
#Aplicando transformacion Logaritmica
df['SalePrice_log'] = np.log(df['SalePrice'] + 1) 
df['YearBuilt_log'] = np.log(df['YearBuilt'] + 1) 
```

```python
#Funcion para graficar la distribucion de una variable/columna antes y despues de las transformaciones logaritmicas
def plot_distribution(original, transformed, column):
    plt.figure(figsize=(14, 6))

    # Gráfico de la variable original
    plt.subplot(1, 2, 1)
    sns.histplot(original, kde=True)
    plt.title(f'Distribución original de {column}')

    # Gráfico de la variable transformada
    plt.subplot(1, 2, 2)
    sns.histplot(transformed, kde=True)
    plt.title(f'Distribución logarítmica de {column}')
    
    plt.show()
```

```python
# Visualizamos la distribución de las columnas con sesgo alto despues de la transformacion logartimica
for col in skewed_columns:
    plot_distribution(df[col], df[col + '_log'], col)
```

```python
#Creando nuevas caracteristicas
df['TotalArea']= df['GrLivArea'] + df['TotalBsmtSF']
df['Price_per_sqft'] = df['SalePrice'] / df['GrLivArea']
print(df[["TotalArea",'Price_per_sqft']].head())
```

```python
#Verficando el valor maximo de la columna Price_per_sqft
print(df['Price_per_sqft'].max())
```

```python
from sklearn.preprocessing import LabelEncoder
df['Price_per_sqft_Category'] = pd.cut(df['Price_per_sqft'], bins=[0, 100, 200, df['Price_per_sqft'].max()], labels=['Bajo', 'Medio', 'Alto'], include_lowest=True)

category_order = ['Bajo', 'Medio', 'Alto']
category_df = pd.DataFrame({'Category': category_order, 'Order': range(len(category_order))})

# Mapeo de las categorías a números basados en el DataFrame auxiliar
category_map = category_df.set_index('Category')['Order'].to_dict()
df['Price_per_sqft_Category_Encoded'] = df['Price_per_sqft_Category'].map(category_map)

# Mostrar el resultado
print(df[['Price_per_sqft_Category', 'Price_per_sqft_Category_Encoded']].head())
```
